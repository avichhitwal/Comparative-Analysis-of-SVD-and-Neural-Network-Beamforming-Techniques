```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from scipy.linalg import svd
import random
```

SYSTEM PARAMETERS

```python
n_tx = 64    # Number of Tx antennas
n_rx = 8     # Number of Rx antennas
mod_order = 4  # QPSK
num_samples = 10000  # Training examples

snr_db_train = 20  # SNR for training (can vary later)
```

CHANNEL AND DATASET GENERATOR

```python
def generate_channel_matrix(n_tx, n_rx):
    """Generate Rayleigh flat fading channel."""
    H_real = np.random.randn(n_rx, n_tx) / np.sqrt(2)
    H_imag = np.random.randn(n_rx, n_tx) / np.sqrt(2)
    H = H_real + 1j * H_imag
    return H

def optimal_beamformer(H):
    """Compute optimal Tx beamforming vector (SVD approach)."""
    U, S, Vh = svd(H)
    f_opt = Vh.conj().T[:, 0]  # First right singular vector
    return f_opt
```

DATASET


```python
X = []
Y = []

for _ in range(num_samples):
    H = generate_channel_matrix(n_tx, n_rx)
    f_opt = optimal_beamformer(H)

    # Add noise (simulate imperfect CSI)
    noise = (np.random.randn(*H.shape) + 1j*np.random.randn(*H.shape)) / np.sqrt(2)
    snr_linear = 10**(snr_db_train/10)
    H_noisy = H + noise / np.sqrt(snr_linear)

    X.append(np.hstack([H_noisy.real.flatten(), H_noisy.imag.flatten()]))
    Y.append(np.hstack([f_opt.real, f_opt.imag]))  # Predict real and imaginary parts

X = np.array(X)
Y = np.array(Y)

# Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```

NEURAL NETWORK

```python
model = keras.Sequential([
    layers.Input(shape=(2 * n_tx * n_rx,)),
    layers.Dense(512, activation='relu'),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Flatten(), # Flatten the output of the previous layer
    layers.Dense(2 * n_tx)  # Real and imaginary parts
])

model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
history = model.fit(X_train, Y_train, epochs=50, batch_size=64, validation_split=0.1)
```

```python
loss = model.evaluate(X_test, Y_test)
print("Test MSE Loss:", loss)
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training/Validation Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()
plt.grid(True)
plt.show()
```

QPSK

```python
def qpsk_modulate(bits):
    bits = bits.reshape(-1, 2)
    mapping = {
        (0,0): 1 + 1j,
        (0,1): -1 + 1j,
        (1,0): 1 - 1j,
        (1,1): -1 - 1j
    }
    symbols = np.array([mapping[tuple(b)] for b in bits])
    symbols /= np.sqrt(2)
    return symbols

def qpsk_demodulate(symbols):
    if np.isscalar(symbols):
        symbols = [symbols]
    bits = np.zeros((len(symbols), 2), dtype=int)
    bits[:, 0] = np.real(symbols) < 0
    bits[:, 1] = np.imag(symbols) < 0
    return bits.flatten()
```

```python
num_bits = 10000
snr_db_test = np.arange(0, 21, 4)  # SNR points from 0 to 20 dB

ber_svd = []
ber_nn = []
```

```python
for snr_db in snr_db_test:
    errors_svd = 0
    errors_nn = 0
    n_blocks = num_bits // (2 * n_rx)

    for _ in range(n_blocks):

        H = generate_channel_matrix(n_tx, n_rx)

        bits = np.random.randint(0, 2, size=(num_bits,))
        symbols = qpsk_modulate(bits)

        # Traditional (SVD)
        U, S, Vh = svd(H)
        f_svd = Vh.conj().T[:, 0]
        f_svd /= np.linalg.norm(f_svd)

        y_svd = H @ f_svd * symbols[0]
        noise = (np.random.randn(*y_svd.shape) + 1j*np.random.randn(*y_svd.shape)) / np.sqrt(2)
        snr_linear = 10**(snr_db/10)
        y_svd_noisy = y_svd + noise/np.sqrt(snr_linear)

        received_symbol_svd = np.vdot(U[:, 0], y_svd_noisy)
        detected_bits_svd = qpsk_demodulate(np.array([received_symbol_svd]))

        detected_bits_svd = detected_bits_svd.flatten()
        detected_bits_svd = np.repeat(detected_bits_svd, num_bits // 2)
        errors_svd += np.sum(bits != detected_bits_svd)

        # Neural Network Beamformer
        H_input = np.hstack([H.real.flatten(), H.imag.flatten()]).reshape(1, 2 * n_tx * n_rx)
        f_nn_pred = model.predict(H_input, verbose=0).flatten()
        f_nn = f_nn_pred[:n_tx] + 1j * f_nn_pred[n_tx:]
        f_nn /= np.linalg.norm(f_nn)

        y_nn = H @ f_nn * symbols[0]
        noise = (np.random.randn(*y_nn.shape) + 1j*np.random.randn(*y_nn.shape)) / np.sqrt(2)
        y_nn_noisy = y_nn + noise/np.sqrt(snr_linear)

        received_symbols_nn = np.vdot(U[:, 0], y_nn_noisy)
        detected_bits_nn = qpsk_demodulate(received_symbols_nn)

        detected_bits_nn = detected_bits_nn.flatten()
        detected_bits_nn = np.repeat(detected_bits_nn, num_bits // 2)


        errors_nn += np.sum(bits != detected_bits_nn)

    # BER
    ber_svd.append(errors_svd / (num_bits))
    ber_nn.append(errors_nn / (num_bits))

    print(f"SNR {snr_db} dB: SVD BER = {ber_svd[-1]:.5f}, NN BER = {ber_nn[-1]:.5f}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(snr_db_test, ber_svd, 'o-', label='SVD Beamforming')
plt.plot(snr_db_test, ber_nn, 's--', label='Neural Network Beamforming')
plt.title('Bit Error Rate (BER) vs SNR for Different Beamforming Techniques')
plt.xlabel('SNR (dB)')
plt.ylabel('Bit Error Rate (BER)')
plt.legend()
plt.grid(True)
plt.yscale('log')
plt.show()
```
